```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
import numpy as np
```

```python
def kernel_matrix(x : torch.Tensor, sigma):
    dim = len(x.size())
      #dim = 1

    x1  = torch.unsqueeze(x, 0)
    x2  = torch.unsqueeze(x, 1)
    axis= tuple(range(2, dim+1))
    if dim > 1:

      result1 = torch.exp( -0.5 * torch.sum(torch.pow(x1-x2, 2), axis=axis) / sigma**2)
      return result1 #RBF kernel
    else:
      result2 = torch.exp( -0.5 * torch.pow(x1-x2, 2) / sigma**2)
      return result2 #RBF kernel

def HSIC(Kx, Ky, m): #finds HSIC value. A larger value indicates strong dependence.
    Kx = Kx.double()
    Ky = Ky.double()
    xy = torch.matmul(Kx, Ky)
    h  = torch.trace(xy) / m**2 + torch.mean(Kx)*torch.mean(Ky) - \
        2 * torch.mean(xy)/m
    return h*(m/(m-1))**2
```
